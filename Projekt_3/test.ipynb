```python
from logic import ImageProcessor
import matplotlib.pyplot as plt
from PIL import Image
```

```python
processor = ImageProcessor()
img = Image.open("w5.bmp")
skeletonize_array = processor.skeletonize(img)
KMM_array = processor.KMM(img)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img.convert('L'), cmap='gray')
axes[0].set_title('Oryginał (w5.bmp)')
axes[0].axis('off')

axes[1].imshow(skeletonize_array, cmap='gray')
axes[1].set_title('Szkieletyzacja (Morfologia)')
axes[1].axis('off')

axes[2].imshow(KMM_array, cmap='gray')
axes[2].set_title('Algorytm KMM')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
